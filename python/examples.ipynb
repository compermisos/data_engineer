```python
class NetworkTimeoutError(Exception):
    def __init__(self, host, port, timeout_limit):
        self.host = host
        self.port = port
        self.timeout_limit = timeout_limit
        # Pass a default message to the base Exception class
        super().__init__(f"Connection to {host}:{port} timed out after {timeout_limit}s")

# Raising the exception with custom attributes
raise NetworkTimeoutError("localhost", 8080, 5)
```

```python
def divide(number: int | float) -> float:
    try:
        resultado = 10 / number
    except ValueError:
        print("Por favor, introduce un valor numérico válido.")
    except ZeroDivisionError:
        print("¡No puedes dividir entre cero!")
    else:
        return resultado

numero = int(input("Introduce un número: "))
print(divide(numero))
print(divide(2.5))
```

```python
def greet(name: str | None = None) -> str:
    if name is None:
        return "Hello, Guest!"
    return f"Hello, {name}!"

print(greet())
print(greet("Christian"))
```

```python
from dataclasses import dataclass

@dataclass
class Persona:
    nombre: str
    edad: int
    activo: bool = True

p1 = Persona("Ana", 28)
p2 = Persona("Ana", 28)
print(p1) # Muestra: Persona(nombre='Ana', edad=28, activo=True)
print(p1 == p2) # Muestra: True
```

```python
from dataclasses import dataclass

@dataclass
class RangoRecursivo:
    valor: int
    minimo: int = 1
    maximo: int = 999
    
    def __post_init__(self):
        # Limita el valor al rango establecido
        self.valor = max(self.minimo, min(self.maximo, self.valor))
```

```python
def divide(number: RangoRecursivo):
    try:
        resultado = 10 / number.valor
        print(number.valor)
    except ValueError:
        print("Por favor, introduce un valor numérico válido.")
    except ZeroDivisionError:
        print("¡No puedes dividir entre cero!")
    else:
        print(resultado)

numero = RangoRecursivo(int(input("Introduce un número: ")))
divide(numero)
```
